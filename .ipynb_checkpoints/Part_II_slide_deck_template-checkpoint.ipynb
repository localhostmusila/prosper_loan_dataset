## Dataset Overview
Loan Data from Prosper data set contains 113,937 loans with 81 variables on each loan, including loan amount, borrower rate (or interest rate), current loan status, borrower income, and many others.

Started with Data wrangling which involved assessing data, data cleaning, manipulation and data visualization which was the main work done on the dataset. Dataset was minised to 16 variables but on analysis LoanOriginationDate variable was divided into loanOrigination_month and loanOrigination_year.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import warnings

from pandas.api.types import CategoricalDtype

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# load in the dataset into a pandas dataframe - saved after cleanup
loan = pd.read_csv('prosperLoanData_master.csv')
```

> Note that the above cells have been set as "Skip"-type slides. That means
that when the notebook is rendered as http slides, those cells won't show up.

```python
#Confirm Dataset Loading
loan.info()
```

```python
#color definition
base_color = sb.color_palette()[0]

base_c1=sb.color_palette()[1]

base_c2=sb.color_palette()[2]
```

```python
# Function for a bar chart (categorical variable)

def barchart_x(i):
    sb.countplot(data = loan, x= i, color=base_color);
    
def barchart_y(i):
    sb.countplot(data = loan, y= i, color=base_color);
```

```python
#Ordering the LoanOriginationQuarter categorical variable
unordered_quarter = ['Q1 2007', 'Q2 2007', 'Q3 2007', 'Q4 2007', 'Q1 2008', 'Q2 2008', 'Q3 2008', 'Q4 2008', 
       'Q1 2009', 'Q2 2009', 'Q3 2009', 'Q4 2009', 'Q1 2010', 'Q2 2010', 'Q3 2010', 'Q4 2010', 
       'Q1 2011', 'Q2 2011', 'Q3 2011', 'Q4 2011', 'Q1 2012', 'Q2 2012', 'Q3 2012', 'Q4 2012',
       'Q1 2013', 'Q2 2013', 'Q3 2013', 'Q4 2013', 'Q1 2014']
ordered_quarter = CategoricalDtype(categories=unordered_quarter, ordered=True)
loan.LoanOriginationQuarter = loan.LoanOriginationQuarter.astype(ordered_quarter)
```

```python
#Sample random 500 loans so that plots are clearer.
np.random.seed(1)
df_sample = loan.sample(500)
```

## BorrowerAPR Distribution

> BorrowerAPR has 3 peaks (multimodal Distribution) This is observed between the regions:
> - Between 0.15 and 0.2
> - At 0.3  
> - Between 0.35 and 0.4

> Other regions remain to be evenly distributed.

```python
#standard-scaled plot
binsize = 0.009
bins = np.arange(0, loan['BorrowerAPR'].max()+binsize, binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = loan, x = 'BorrowerAPR', bins = bins)
plt.xlabel('BorrowerAPR (%)')
plt.ylabel('Count')
plt.show()
```

## BorrowerRate Distribution

> Borrower Interest Rate has a Multimodal Distribution almost similar to that of the BorrowerAPR. This indicates that there is a relationship between this two variables.

```python
#standard-scaled plot
binsize = 0.009
bins = np.arange(0, loan['BorrowerRate'].max()+binsize, binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = loan, x = 'BorrowerRate', bins = bins)
plt.xlabel('BorrowerRate (%)')
plt.ylabel('Count')
plt.show()
```

## LoanOriginationQuarter, loanOrigination_year, loanOrigination_month
> More loan borrowers in 2013 as compared to other year. 2009 had the lowest loan borrowers. 

> We have more borrowers in the months of October, November, December and January. Less borrowers in the month of April.

```python
plt.figure(figsize=[20,8])
plt.subplot(1,3,1)
barchart_y('LoanOriginationQuarter')

plt.subplot(1,3,2)
barchart_x('loanOrigination_year')

plt.subplot(1,3,3)
barchart_x('loanOrigination_month')
plt.show();
```

## LoanOriginalAmount, Term vs BorrowerRate

- BorrowerRate is maximum at 36 months.
- Most of the borrowers took loans for 36 months.
- Minimum BorrowerRate is approximately 0.05 and maximum BorrowerRate is approximately 0.35

```python
plt.figure(figsize=[20,13])

#Creating ScatterPlot for LoanOriginalAmount, BorrowerRate
plt.subplot(2,2,1)
sb.regplot(data=df_sample, y='BorrowerRate', x='LoanOriginalAmount')
plt.ylabel('Borrower Interest Rate')
plt.xlabel('Loan Original Amount')
plt.title('ScatterPlot of BorrowerRate vs LoanOriginalAmount');

#Creating Heatmap for LoanOriginalAmount, BorrowerRate
plt.subplot(2,2,2)
plt.hist2d(data=df_sample,  y='BorrowerRate', x='LoanOriginalAmount', cmap='viridis_r', cmin=0.5)
plt.ylabel('Borrower Interest Rate')
plt.xlabel('Loan Original Amount')
plt.title('Heatmap of BorrowerRate vs LoanOriginalAmount');
plt.colorbar();

#Creating Violin plot for Term, BorrowerRate
plt.subplot(2,2,3)
sb.violinplot(data=loan, y='BorrowerRate', x='Term',color=base_color)
plt.ylabel('Borrower Interest Rate ')
plt.xlabel('Term (Months)')
plt.title('Violin Plot of BorrowerRate vs Term');

#Creating Boxplot for Term, BorrowerRate
plt.subplot(2,2,4)
sb.boxplot(data=loan,  y='BorrowerRate', x='Term',color=base_color)
plt.ylabel('Borrower Interest Rate')
plt.xlabel('Term (Months)')
plt.title('Box Plot of BorrowerRate vs Term');
```

## LoanOriginalAmount vs BorrowerAPR

- BorrowerAPR count highest between 1000 and 7000.
- We have a negative correlation between BorrowerAPR and LoanOriginalAmount. As LoanOriginalAmount increases, BorrowerAPR is decreasing and BorrowerAPR count is decreasing significantly after 16,000.
- When LoanOriginalAmount between 1000 - 7000 highest count of BorrowerAPR is 6706 at approximately 0.35.

It can be noted that when LoanOriginalAmount is small it has more borrowers and very high BorrowerAPR.

```python
#Plotting Influence of LoanOriginalAmount on BorrowerAPR (Heatmap)
plt.figure(figsize=(20,8))
bins_x = np.arange(1000,35000+5000,3000)
bins_y=np.arange(0.05,0.40+0.06,0.04)
labels = ['{}'.format(x) for x in bins_x]
gt = plt.hist2d(y='BorrowerAPR',x='LoanOriginalAmount',data=loan,cmin=0.5,cmap='viridis_r',bins=[bins_x,bins_y])
count = gt[0]
for i in range(count.shape[0]):
    for j in range(count.shape[1]):
        annot = count[i,j]
        if annot!=np.nan and annot >= 2000:
            plt.text(x=gt[1][i]+50,y=gt[2][j]+0.005,s=int(annot),color='white')
        elif annot !=np.nan and annot > 0:
             plt.text(x=gt[1][i]+50,y=gt[2][j]+0.005,s=int(annot),color='black')
plt.colorbar()
plt.xticks(bins_x,labels)
plt.title('Influence of LoanOriginalAmount on BorrowerAPR (Heatmap)')
plt.xlabel('LoanOriginalAmount')
plt.ylabel('BorrowerAPR');
```

## BorrowerAPR, LoanStatus, Term

- Defaulted Status: Loans with Term of 12months have the highest BorrowerAPR.
- All other LoanStatus with a Loan Term of 12months show high BorrowerAPR compared to Term 36 and 60 months.

```python
plt.figure(figsize=[20,10])
sb.boxplot(data = loan, y = 'BorrowerAPR', x = 'LoanStatus', hue = 'Term')
plt.legend(loc = 6, bbox_to_anchor = (1, 0.3))
plt.xticks(rotation = 80);
```
